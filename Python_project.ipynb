```python
COL.set_index('Date',inplace=True)
W.set_index('ObsDate',inplace=True)
COW=COL.merge(W,left_on='Date',right_index=True)
```

```python
COW['Total Hits']=COW['Home Hits']+COW['Visitor Hits']
COW['Total Score']= COW['Home Score']+COW['Visitor Score']
COW['Hits div Score']=COW['Total Hits']/COW['Total Score']
```
